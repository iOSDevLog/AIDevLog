Jupyter笔记本有两种不同的键盘输入模式：

* **编辑模式** 允许您将代码或文本输入到一个单元格中，并通过一个绿色的单元格来表示 
* **命令模式** 将键盘与笔记本级命令绑定在一起，并通过一个灰色的单元格边界显示，该边框为蓝色的左边框

## Mac OS X 快捷键说明

| 快捷键 | 说明  |
| ------ | ----- |
| ⌘      | Cmd 命令 |
| ⌃      | Ctrl 控制 |
| ⌥      | Option 选项 |
| ⇧      | Shift |
| ↩      | 返回  |
| ␣      | 空格  |
| ⇥      | Tab   |
